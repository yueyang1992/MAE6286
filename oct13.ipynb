```python
import numpy                       
from matplotlib import pyplot                 
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
nx = 41  # try changing this number from 41 to 81 and Run All ... what happens?
dx = 2/(nx-1)
nt = 25
sigma = 0.2
c = 1
dt = sigma*dx/c      
x = numpy.linspace(0,1,nx)
```

```python
u = numpy.zeros(nx)
for i in range(nx):
    if 0.5 <= x[i]: 
        u[i] = 0
    if 0.5 >= x[i]:
        u[i] = 1
```

```python
pyplot.plot(x, u, color='#003366', ls='--', lw=3)
pyplot.ylim(-0.5,1.5);
```

```python
for n in range(nt):  
    un = u.copy() 
    u[1:-1] = 0.5*(un[2:]+un[0:-2])-0.5*c*dt/dx*(un[2:]-un[0:-2])
```

```python
pyplot.plot(x, u, color='#003366',ls='--', lw=3)
pyplot.ylim(-0.5,1.5)
pyplot.xlim(0,1)
```

```python
for n in range(nt):  
    un = u.copy() 
    u[1:-1] = un[1:-1]-0.5*sigma*(un[2:]+un[0:-2])-0.5*sigma**2*(un[2:]-un[1:-1]+un[0:-2])
```

```python
pyplot.plot(x, u, color='#003366',ls='--', lw=3)
pyplot.ylim(-0.5,1.5)
pyplot.xlim(0,1)
```

```python
for n in range(nt):  
    un = u.copy() 
    u[1:-1] = un[1:-1]-c*dt/dx*(un[2:]-un[0:-2])
```

```python
pyplot.plot(x, u, color='#003366',ls='--', lw=3)
pyplot.ylim(-0.5,1.5)
pyplot.xlim(0,1)
```
